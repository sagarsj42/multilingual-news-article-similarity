```python
filename1 = 'xlmr_test_results.csv'
filename2 = 'distilbert-metadata-results.csv'
```

```python
import pandas as pd
```

```python
pair_ids = pd.read_csv(filename1).pair_id.tolist()
pair_ids
```

```python
preds1 = pd.read_csv(filename1).Overall.values
preds1
```

```python
preds2 = pd.read_csv(filename2).Overall.values
preds2
```

```python
preds = (preds1 + preds2) / 2
preds.tolist()
```

```python
pd.DataFrame(list(zip(pair_ids, preds)), columns=['pair_id', 'Overall']).to_csv('average-db-xlmr-test-results.csv')
```
